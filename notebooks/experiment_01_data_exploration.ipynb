```python
import nibabel as nib
import matplotlib.pyplot as plt
import os

# Updated filename
file_path = "../data/raw/brain_mri.nii.gz"

if os.path.exists(file_path):
    img = nib.load(file_path)
    data = img.get_fdata()
    print(f"✅ Loaded Brain Scan with shape: {data.shape}")
    
    # Show the middle slice (Sagittal view)
    # The MNI brain is usually around shape (91, 109, 91)
    mid_slice_x = data.shape[0] // 2
    
    plt.imshow(data[mid_slice_x, :, :].T, cmap="gray", origin="lower")
    plt.axis("off")
    plt.title(f"Mid-Sagittal Slice (x={mid_slice_x})")
    plt.show()
else:
    print("❌ File not found. Did the download script finish successfully?")
```
