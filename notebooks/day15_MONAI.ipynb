```python
import monai
print(monai.__version__)
```

```python
import torch
from monai.transforms import Compose, ScaleIntensity
from monai.networks.nets import UNet
```

```python
# create a dummy medical image 3D
# simulated 3D medical image (1 channel, 64*64*64)
image = torch.rand(1, 64, 64, 64) # (channels, depth, height, width)
print(image.shape)
```

```python
# apply MONAI transform
transform = Compose([
    ScaleIntensity()
])

image_transformed = transform(image)
```

```python
# define a MONAI UNet
model = UNet(
    spatial_dims=3,
    in_channels=1,
    out_channels=2,
    channels=(16, 32, 64),
    strides=(2,2),
)
```

```python
# forward pass (hello world moment)
output = model(image_transformed.unsqueeze(0))
print(output.shape)  # (batch, classes, depth, height, width)
```
