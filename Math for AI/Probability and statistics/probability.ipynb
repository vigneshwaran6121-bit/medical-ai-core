```python
# Noisy sensor data
import numpy as np
import matplotlib.pyplot as plt

# Stimulate : AI reading a patient's temperature 1000 times (noisy sensor)
# True temp is 37.0, but the sensor has noise (std_dev of 0.5)
data = np.random.normal(loc=37.0, scale=0.5, size=1000)

mean_val = np.mean(data)
std_dev = np.std(data)

print(f"AI Estimate(Mean): {mean_val:.2f}")
print(f"Uncertainty (Std Dev): {std_dev:.2f}")

# if we see a new reading of 40.0 is it fever or noise?
# 40.0 is (40 - 37) / 0.5 = 6 standard deviation away
# it is DEFINITELY fever, not noise
```

Baye's Theorem

```python
def bayes_update(prior, sensitivity, false_positive_rate):
    # P(Disease)
    p_disease = prior

    # p(positive Test / Disease)
    p_pos_given_disease = sensitivity

    # P (positive Test / Disease)
    p_pos_given_healthy = false_positive_rate

    # P(No Disease)
    p_healthy = 1 - prior

    # total probability of a positive test
    p_pos = (p_pos_given_disease * p_disease) + (p_pos_given_healthy * p_healthy)

    # Bayes formula: p(Disease / Positive Test)
    posterior = (p_pos_given_disease * p_disease) / p_pos 
    return posterior

#scenario : rate disease (1% of pop). Test is 90% accurate
#false positive rate is 9%
result = bayes_update(prior= 0.01, sensitivity= 0.90, false_positive_rate=0.09)

print(f"Probability of disease after positive test: {result:.2%}")
# Result is often shockingly low (around 9%), proving why doctors double-check!
```
