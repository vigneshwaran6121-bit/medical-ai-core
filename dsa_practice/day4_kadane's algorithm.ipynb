```python
def maxsubarray(nums):
    global_max=nums[0]
    local_max=nums[0]

    for num in nums[1:]:
        local_max = max(num, num+local_max)
        global_max = max(global_max, local_max)

    return global_max
```

```python
nums = [5, -3, 4, -2, 1]
ans = maxsubarray(nums)
print ("max sub array is :", ans)
```
