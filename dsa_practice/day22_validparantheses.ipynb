```python
def isvalid(s):
    stack = []
    mapping = {')': '(', '}': '{', ']': '['}

    for ch in s:
        if ch in mapping:
            if not stack:
                return False
            
            top = stack.pop()
            print("top", top)
            #print("stack", stack)
            if top!= mapping[ch]:
                return False
        else:
            stack.append(ch)

    return len(stack) == 0
print(isvalid("()[]{}"))  # True
#print(isvalid("(]"))      # False
#print(isvalid("([)]"))    # False
#print(isvalid("{[]}"))    # True
```
