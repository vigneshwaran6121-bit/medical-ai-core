```python
def length_of_longest_substring(s: str):
    char_set = set()
    left = 0
    max_length = 0

    for right in range(len(s)):
        # if the character at 'right' is already in our window, 
        # shrink the window from the left until it's removed
        while s[right] in char_set:
            char_set.remove(s[left])
            left += 1

            # add the current chara cter to the set
            char_set.add(s[right])

            # update our result
            # (right - left +  1) is the current size of the window
            max_length = max(max_length, right - left + 1)

        return max_length
```

```python
# test case
# inpur : "abcabcbb"
# explanation: the answer is "abc", with the length of 3
print(length_of_longest_substring("abcabcbb"))
```
