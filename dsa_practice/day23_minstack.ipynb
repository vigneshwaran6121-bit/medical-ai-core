```python
class minstack:
    def __init__(self):
        self.stack = []
        self.min_stack = []

    def push(self, val):
        self.stack.append(val)
        if not self.min_stack:
            self.min_stack.append(val)
        else:
            self.min_stack.append(min(val, self.min_stack[-1]))

    def pop(self):
        self.stack.pop()
        self.min_stack.pop()

    def top(self):
        return self.stack[-1]
    
    def getmin(self):
        return self.min_stack[-1]
    
minstack = minstack()

minstack.push(5)
print("stack:", minstack.stack)
print("min_stack:", minstack.min_stack)
print("current min:", minstack.getmin())
print()

minstack.push(2)
print("stack:", minstack.stack)
print("min_stack:", minstack.min_stack)
print("current min:", minstack.getmin())
print()

minstack.push(8)
print("stack:", minstack.stack)
print("min_stack:", minstack.min_stack)
print("current min:", minstack.getmin())
print()

minstack.push(1)
print("stack:", minstack.stack)
print("min_stack:", minstack.min_stack)
print("current min:", minstack.getmin())
```
