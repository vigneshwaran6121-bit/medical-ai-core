```python
def max_profit(prices: list[int]):
    min_price = float('inf') # start with a very high number
    max_profit = 0

    for price in prices:
        if price < min_price:
            # update our "buy" price if we find something cheaper
            min_price = price
        elif(price - min_price) > max_profit:
            # calculate profit for selling today and check if it's the best
            max_profit = price - min_price

    return max_profit
```

```python
# test case
# input : [7, 1, 5, 3, 6, 4]
# explanation : buy at 1(day 2), sell at 6(day 5). Profit = 5
print (max_profit([7, 1, 5, 3, 6, 4]))
```
