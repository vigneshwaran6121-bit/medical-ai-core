```python
def productexceptself(nums):
    n=len(nums)
    output = [1]*n
    #print(output)
    
    left_product = 1
    for i in range (n):
        output[i] = left_product
        left_product *= nums[i]
        print(left_product)

    right_product = 1
    for i in range(n-1, -1, -1):
        #print(i)
        output[i]*= right_product
        right_product *= nums[i]
        print(right_product)
    
    return output
```

```python
nums = [1,2,3,4]
print(productexceptself(nums))
```

```python
def product(nums):
    n= len(nums)
    out = [1]*n
    #return out

    right=1
    for i in range(n):
        print(f'right before = {right}')
        out[i] = right
        print(f"out = {out}")
        right*=nums[i]
        print(f"right={right}")
    return right

nums=[1, 2, 3, 4, 5]
print(product(nums))
```
