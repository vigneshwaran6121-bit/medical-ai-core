```python
def checkinclusion(s1: str, s2: str) -> bool:
    if len(s1)> len(s2):
        return False
    
    freq1 = {}
    freq2 = {}

    # build freq map for s1
    for char in s1:
        freq1[char] = freq1.get(char, 0) + 1

    left = 0
    window_size = len(s1)

    for right in range(len(s2)):
        # add current char to window map
        freq2[s2[right]] = freq2.get(s2[right], 0) + 1

        # keep window size fixed
        if right - left + 1 > window_size:
            freq2[s2[left]] -= 1
            if freq2[s2[left]] == 0:
                del freq2[s2[left]]
            left += 1

        # compare freq maps
        if freq1 == freq2:
            return True
        
    return False

s1 = "ab"
s2 = "eidbaooo"
print(checkinclusion(s1, s2))

```

```python
def duplicate(nums):
    seen={}
    for num in nums:
        if num in seen:
            return True
        seen[num]=1
    return False


nums=[1,2,3,1]
print(duplicate(nums))
```

```python
def ana(s1, s2):
    if len(s1)!=len(s2):
        return False
    
    freq={}

    for char in s1:
        if char in freq:
            freq[char] +=1
        else:
            return False
    for char in s2:
        freq[char]=freq.get(char,0)-1
    
    for count in freq.values():
        if count !=0:
            return False
    return True
s1 = "rat"
s2 = "tar"
print(ana(s1, s2))
```

```python
def longestconsecutive(nums):
    num_set = set(nums)
    print(num_set)
    longest = 0

    for num in num_set:
        #check if num is start of sequence
        if num-1 not in num_set:
            current = num
            length = 1

            while current +1 in num_set:
                current +=1
                length +=1

            longest = max(longest, length)

    return longest

s = [100, 4, 200, 1, 3, 2]
print(longestconsecutive(s))
```

```python
def ispalindrom(s: str):
    left = 0
    right = len(s)-1
    #print(right)

    while left < right:
        # skip non-alphanumeric charac
        while left < right and not s[left].isalnum():
            left += 1
        while left < right and not s[right].isalnum():
            right -= 1

        # compare charac
        if s[left].lower() != s[right].lower():
            return False
        
        left +=1
        right -=1

    return True

        
```

```python
print(ispalindrom("A man, a plan, a canal: Panama"))  # True
print(ispalindrom("race a car"))                      # False
print(ispalindrom(" "))                               # True
print(ispalindrom("madam"))                           # True
print(ispalindrom("No lemon, no melon"))              # True
```

```python
def polindrom(s:str):
  left = 0
  right = len(s)-1

  while left<right:
    while left<right and not s[left].isalnum():
      left += 1
    while left<right and not s[right].isalnum():
      right -=1

    if s[left].lower != s[right].lower:
      return False
    left  += 1
    right -=1
  return True

print(polindrom("A man, a plan, a canal: Panama"))  # True
print(polindrom("race a car"))                      # False
print(polindrom(" "))                               # True
print(polindrom("madam"))                           # True
print(polindrom("No lemon, no melon"))  
```
