```python
def dailytemp(temp):
    n= len(temp)
    answer = [0]*n
    stack=[]

    for i in range(n):
        while stack and temp[i] > temp[stack[-1]]:
            prev = stack.pop()
            answer[prev] = i-prev
        stack.append(i)

    return answer

# Example usage:
temperatures = [73, 74, 75, 71, 69, 72, 76, 73]
print(dailytemp(temperatures))  # Output: [1, 1,
```
