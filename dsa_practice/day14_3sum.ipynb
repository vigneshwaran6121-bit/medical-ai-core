```python
def three_sum(nums: list[int]):
    nums.sort() # sorting is mandatory for this approach : O(N log N)
    print("sorted nums:", nums)
    result = []
    n= len(nums)

    for i in range(n-2):
        # skip duplicates for the 1st number(i)
        if i>0 and nums[i] == nums[i-1]:
            continue

        left = i + 1
        right = n-1

        while left<right:
            total = nums[i] + nums[left] + nums[right]

            if total < 0:
                # sum is too small, need bigger numbers --> move left forward
                left += 1
            elif total > 0:
                # sum is too big, need smaller numbers --> move right forward
                right -= 1
            else:
                # we found a triplet!
                result.append([nums[i], nums[left], nums[right]])

                # skip duplicates for the second number to ensure unique triplets
                while left < right and nums[left] == nums[left+1]:
                    left += 1
                
                # move both pointers inward
                left += 1
                right -= 1

    return result
    
    
```

```python
# test case
#input:[-1, 0, 1, 2, -1, -4]
#sorted: [-4, -1, -1, 0, 1, 2]
print(three_sum([-1, 0, 1, 2, -1, -4]))
#output: [[-1, -1, 2], [-1, 0, -1]]
```
