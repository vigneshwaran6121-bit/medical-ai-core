```python
# longest consecutive integer

def LongestConsecutive(nums):
    num_set = set(nums)
    print(num_set)
    longest = 0

    for num in num_set:
        # check if num is start of sequence
        if num - 1 not in num_set:
            current = num
            length = 1
            #print("\n1st current:", current, "1st length:", length)

            while current +1 in num_set:
                current += 1
                length += 1
            #print("2nd longest:", longest, "2nd length:", length)
            longest = max(longest, length)
            #print("3rd longest:", longest, length, "3rd current:", current)

    return longest
nums = [100, 4, 200, 1, 3, 2]
result = LongestConsecutive(nums)
print ("\nThe Longest Integer is :", result)
```
