```python
def evalrpn(tokens):
    stack=[]

    for token in tokens:
        if token not in "+-*/":
            stack.append(int(token))
        else:
            b=stack.pop()
            a=stack.pop()

            if token=="+":
                stack.append(a+b)
            elif token=="-":
                stack.append(a-b)
            elif token=="*":
                stack.append(a*b)
            else:
                stack.append(int(a/b))
    return stack[0]
print(evalrpn(["2","1","+","3","*"]))  # Output: 9
```
