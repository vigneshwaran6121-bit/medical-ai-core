```python
def max_area(height: list[int]):
    left = 0
    right = len(height)-1
    max_water = 0

    while left<right:
        # calculate height of the water level (limited by the shorter line)
        h = min(height[left], height[right])
        w = right - left

        # update max area
        max_water = max(max_water, h*w)

        # move the pointer of the shorter line inward
        if height[left] < height[right]:
            left += 1
        else:
            right -= 1

    return max_water
```

```python
# test case
# input : [1, 8, 6, 2, 5, 4, 8, 3, 7]
# Explanation : the max area is between the 8 (index i) and 7 (index 8)
# width = 7, Height = 7, area = 49
print("area of container:", max_area([1, 8, 6, 2, 5, 4, 8, 3, 7])) # output: 49
```
