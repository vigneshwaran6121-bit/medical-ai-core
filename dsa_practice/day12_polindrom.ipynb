```python
def is_palindrom(s: str):
    left = 0
    right = len(s) - 1

    while left < right:
        # move left pointer forward intil we find a valid char
        if not s[left].isalnum():
            left += 1
            continue

        # move right pointer backward untill we find a valid char
        if not s[right].isalnum():
            right -= 1
            continue

        # compare characters (case - insensitive)
        if s[left].lower() != s[right].lower():
            return False
        
        # update pointers after a succesful match
        left += 1
        right -= 1

    return True

# test class
print(is_palindrom("A man, a plan, a canal: Panama")) # True
print(is_palindrom("race a car"))  # False
```
